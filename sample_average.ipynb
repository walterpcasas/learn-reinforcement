```python
# Generamos los valores verdaderos q*(a) para 10 acciones
q_star = np.random.normal(loc=0.0, scale=1.0, size=10)

# Creamos un grfico con las distribuciones normales para cada accin
x = np.linspace(-4, 4, 500)
fig, ax = plt.subplots(figsize=(10, 6))

for i, q in enumerate(q_star):
    # Creamos la distribucin normal N(q, 1)
    y = (1/np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - q)**2))
    ax.plot(x, y + i, color='gray')  # elevamos cada curva para que no se superpongan
    ax.text(q, i + 0.4, f'q*({i})={q:.2f}', fontsize=9, ha='center')

ax.set_title('Ejemplo de un problema del testbed de 10 bandits')
ax.set_xlabel('Recompensa')
ax.set_ylabel('Accin (desplazada en el eje Y para visualizacin)')
ax.grid(True)
plt.tight_layout()
plt.show()
```
